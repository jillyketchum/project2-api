```python
#download the movie posters for 10 of these movies and save them to this repository
import requests
import os
import pandas as pd

url = "https://api.themoviedb.org/3/discover/movie?include_adult=false&language=en-US&page=1&sort_by=popularity.desc"
headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiIzMjliOWM1OGQ5NjgxNjJjNTViMjZhNTNhYmRkOWI5NSIsIm5iZiI6MTcyNzI5MzQ4OC4xMzY4MTcsInN1YiI6IjY2ZjQyMmI2Njg2MjhhMmVjMDBkZjE5OCIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.LKsnOJsM0hR7MI3_rFkG8D4tRqYPkqZ4GYssOFaEzr8"
}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    data = response.json()
    movies_df = pd.DataFrame(data['results'])  
    print(movies_df.shape)
    print(movies_df.head(5)) 

    poster_base_url = "https://image.tmdb.org/t/p/w500"
    poster_directory = "movie_posters"
    if not os.path.exists(poster_directory):
        os.makedirs(poster_directory)

    for idx, movie in movies_df.head(10).iterrows():
        poster_path = movie['poster_path']
        
        if poster_path:
            poster_url = poster_base_url + poster_path
            img_response = requests.get(poster_url, stream=True)

            if img_response.status_code == 200:
                img_filename = os.path.join(poster_directory, f"movie_{idx + 1}.jpg")
                with open(img_filename, 'wb') as img_file:
                    for chunk in img_response.iter_content(1024):
                        img_file.write(chunk)
                print(f"Downloaded poster for {movie['title']}")
            else:
                print(f"Failed to download poster for {movie['title']}")
else:
    print("Failed to retrieve movie data from TMDb API")
```
